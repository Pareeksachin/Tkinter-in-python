```python
# counting the second programe from tkinter

from tkinter import *
import tkinter as tk
counter=0
def count(label):
    counter=0
    def count1():
        global counter
        counter+=1
        label.configure(text=str(counter))
        label.after(10,count1)
    count1()

    
app=Tk()   
app.title("Counting Seconds")
label=Label(app,fg='Black').pack()
count(label)
button=Button(app,text="Stop",width=25,command=app.destroy).pack()
app.mainloop()
```

```python
# pop menu
from tkinter import *
app=Tk()
label=Label(app,text="click here to show pop up menu",height=20,width=40).pack()

popup=Menu(app,tearoff=0)
popup.add_command(label="Copy")
popup.add_command(label="Paste")
popup.add_command(label="Cut")
popup.add_separator()
popup.add_command(label="Quit")

def do_popup(event):
    try:
        popup.tk_popup(event.x_app,event.y_app,0)
    finally:
        popup.grab_release()


app.bind("<Button - 3>",do_popup)
b=Button(app,text="quit",command=app.quit).pack()
app.mainloop()
```

```python
from tkinter import*
import os

creds='credss.temp'


def signup():
    global password
    global nameL
    global app
    
    app=Tk()
    app.title("signup")
    instruct=Label(app,text="please enter the username and password\n")
    instruct.grid(row=0,column=0,sticky=E)
    nameL=Label(app,text="User Name")
    password=Label(app,text="password")
    nameL.grid(row=1,column=1)
    password.grid(row=2,column=1)
    
    nameL=Entry(app)
    password=Entry(app,show="*")
    nameL.grid(row=1,column=1)
    password.grid(row=2,column=1,sticky=W)
    
    signupbtn=Button(app,text="Sign Up",command=Fsignup)
    signupbtn.grid(columnspan=2,sticky=W)
    app.mainloop()
    
    
def Fsignup():
    with open(creds,'w') as f:
        f.write(nameL.get())
        f.write('\n')
        f.write(passwords.get())
        f.close()
    app.destroy()
    login()
    

def login():
    global nameLL
    global passwordL
    global appl
    
    
    appl=Tk()
    appl.title("Login")
    instructs=Label(appl,text='Login\n')
    instructs.grid(sticky=E)
    nameLL=Label(appl,text="please enter the username")
    passwordL=Label(appl,text="please enter your password")
    nameLL.grid(row=1,column=0)
    passwordL.grid(row=2,column=0)
    
    nameLL=Entry(appl)
    passwordL=Entry(appl,show='*')
    nameLL.grid(row=1,column=1)
    passwordL.grid(row=2,column=1)
    
    
    loginbtn=Button(appl,text="Login",command=Checklogin)
    loginbtn.grid(column=2,sticky=W)
    
    usera=Button(appl,text="Deleteuser",fg='Blue',command=Deluser)
    usera.grid(columnspan=2,sticky=W)

    appl.mainloop()
    
    
def Checklogin():
    with open(creds) as f:
        data=f.readlines()
        uname=data[0].rstrip()
        passwd=data[1].rstrip()
        if nameLL.get()==uname and passwordL.get()==passswd:
            r=Tk()
            r.title("Login Page")
            r.geometry('800x800')
            rlbl=Label(r,text='\n[+] Loggin')
            rlbl.pack()
            r.mainloop()
        
        else:
            r=Tk()
            r.title("login checked")
            r.geometry('800x800')
            rlbl=Label(r,text='\n[+] invalid Loggin')
            rlbl.pack()
            r.mainloop()
        
    
def Deluser():
    os.remove(creds)
    appl.destroy()
    signup()
    
    if os.path.isfile(creds):
        login()
    else:
        signup()
```
